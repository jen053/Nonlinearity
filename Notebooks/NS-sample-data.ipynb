## NS3

```python
# Initial Data import
dfO1_i = pd.read_csv(path+'NS3/Initial/O1/2020-11-18/Trial1/10.0V/Delays.txt', delimiter=' ', header=None, names=
                     ['Delays', 'Time Delays 1', 'Time Delays 2', 'Time Delays 3', 'Time Delays 4'])
dfO2_i = pd.read_csv(path+'NS3/Initial/O2/2020-11-19_02/Trial1/10.0V/Delays.txt', delimiter=' ', header=None, names=
                     ['Delays', 'Time Delays 1', 'Time Delays 2', 'Time Delays 3', 'Time Delays 4'])

# Fracture Data import
dfO1_f = pd.read_csv(path+'NS3/Fracture/O1/2020-11-23_02/Trial1/10.0V/Delays.txt', delimiter=' ', header=None, names=
                     ['Delays', 'Time Delays 1', 'Time Delays 2', 'Time Delays 3', 'Time Delays 4'])
dfO2_f = pd.read_csv(path+'NS3/Fracture/O2/2020-11-24_02/Trial1/10.0V/Delays.txt', delimiter=' ', header=None, names=
                     ['Delays', 'Time Delays 1', 'Time Delays 2', 'Time Delays 3', 'Time Delays 4'])

# Concatinate all NS3 data in one data frame
dfNS3 = pd.concat([dfO1_i['Delays'], dfO1_i['Time Delays 1'], dfO2_i['Time Delays 2'], dfO1_f['Time Delays 3'], 
                  dfO2_f['Time Delays 4']], axis=1)
dfNS3 = dfNS3.rename(columns={"Time Delays 1": "Initial; O1", "Time Delays 2": "Initial; O2", "Time Delays 3": "Post; O1", 
                            "Time Delays 4": "Post; O2"}, errors="raise")
dfNS3
```

## NS5

```python
# Initial Data import
dfO1_i = pd.read_csv(path+'NS5/Initial/O1/2020-11-25/wiggles.txt', delimiter=' ', header=None, names=
                     ['Time Delays 1'])
dfO2_i = pd.read_csv(path+'NS5/Initial/O2/Run1.txt', delimiter=' ')

# Fracture Data import
dfO1_f = pd.read_csv(path+'NS5/Fracture/O1/2020-12-07/Trial1/10.0V/Delays.txt', delimiter=' ', header=None, names=
                     ['Delays', 'Time Delays 1', 'Time Delays 2', 'Time Delays 3', 'Time Delays 4'])
dfO2_f = pd.read_csv(path+'NS5/Fracture/O2/2020-12-07_02/Trial1/10.0V/Delays.txt', delimiter=' ', header=None, names=
                     ['Time Delays 4'])

# Concatinate all NS3 data in one data frame
dfNS5 = pd.concat([dfO1_f['Delays'], dfO1_i['Time Delays 1'], dfO2_i['Time-delay2'], dfO1_f['Time Delays 3'], 
                  dfO2_f['Time Delays 4']], axis=1)
dfNS5 = dfNS5.rename(columns={"Time Delays 1": "Initial; O1", "Time-delay2": "Initial; O2", "Time Delays 3": "Post; O1", 
                            "Time Delays 4": "Post; O2"}, errors="raise")
dfNS5
```

```python
fig, ax = plt.subplots(3,2, figsize=(16,10), sharey=True, sharex=True)
ax[0,0].plot(dfNS3['Delays'], dfNS3['Initial; O1'], '-x', linewidth=2, color='b')
ax[0,0].plot(dfNS3['Delays'], dfNS3['Post; O1'], '-x', linewidth=2, color='r')

ax[0,1].plot(dfNS3['Delays'], dfNS3['Initial; O2'], '-v', linewidth=2, color='b')
ax[0,1].plot(dfNS3['Delays'], dfNS3['Post; O2'], '-v', linewidth=2, color='r')

ax[1,0].plot(dfNS4['Delays'], dfNS4['Initial; O1'], '-x', linewidth=2, color='b')
ax[1,0].plot(dfNS4['Delays'], dfNS4['Post; O1'], '-x', linewidth=2, color='r')

ax[1,1].plot(dfNS4['Delays'][:94], dfNS4['Initial; O2'][:94], '-v', linewidth=2, color='b')
ax[1,1].plot(dfNS4['Delays'], dfNS4['Post; O2'], '-v', linewidth=2, color='r')

ax[2,0].plot(dfNS5['Delays'], dfNS5['Initial; O1'], '-x', linewidth=2, color='b', label='Initial; O1')
ax[2,0].plot(dfNS5['Delays'], dfNS5['Post; O1'], '-x', linewidth=2, color='r', label='Post workflow; O1')

ax[2,1].plot(dfNS5['Delays'], dfNS5['Initial; O2'], '-v', linewidth=2, color='b', label='Initial; O2')
ax[2,1].plot(dfNS5['Delays'], dfNS5['Post; O2'], '-v', linewidth=2, color='r', label='Post workflow; O2')

ax[0,0].set_xlim(-5.0, 100)
ax[0,0].set_ylim(-5.0, 10)
ax[2,1].legend(loc='lower right')
ax[2,0].legend(loc='lower right')
[ax[i, 0].text(5.0, 7.5, l, bbox=dict(boxstyle="round", facecolor='white', alpha=0.5)) 
 for i,l in zip(range(3), ['NS3', 'NS4', 'NS5'])]

ax[1,0].set_ylabel('Time Delay [ns]', fontsize=20)
fig.text(0.5, -0.02, 'Transmission Delay [$\mu$s]', ha='center', fontsize=20)

for i in range(3):
    [ax[i,j].grid() for j in range(2)]
    [ax[i,j].hlines(0, -5.0, 100.0, alpha=0.5) for j in range(2)]
    [ax[i,j].vlines(0, -5.0, 10.0, alpha=0.5) for j in range(2)]
plt.tight_layout()
plt.savefig('c:/users/Jacob/Nonlinearity/Images/Final_results/NS_plot.png', dpi=500, bbox_inches='tight',
           format='png')
plt.show()
```
