## NS3

```python
# Initial Data import
dfO1_i = pd.read_csv(path+'NS3/Initial/O1/2020-11-18/Trial1/10.0V/Delays.txt', delimiter=' ', header=None, names=
                     ['Delays', 'Time Delays 1', 'Time Delays 2', 'Time Delays 3', 'Time Delays 4'])
dfO2_i = pd.read_csv(path+'NS3/Initial/O2/2020-11-19_02/Trial1/10.0V/Delays.txt', delimiter=' ', header=None, names=
                     ['Delays', 'Time Delays 1', 'Time Delays 2', 'Time Delays 3', 'Time Delays 4'])

# Fracture Data import
dfO1_f = pd.read_csv(path+'NS3/Fracture/O1/2020-11-23_02/Trial1/10.0V/Delays.txt', delimiter=' ', header=None, names=
                     ['Delays', 'Time Delays 1', 'Time Delays 2', 'Time Delays 3', 'Time Delays 4'])
dfO2_f = pd.read_csv(path+'NS3/Fracture/O2/2020-11-24_02/Trial1/10.0V/Delays.txt', delimiter=' ', header=None, names=
                     ['Delays', 'Time Delays 1', 'Time Delays 2', 'Time Delays 3', 'Time Delays 4'])

# Concatinate all NS3 data in one data frame
dfNS3 = pd.concat([dfO1_i['Delays'], dfO1_i['Time Delays 1'], dfO2_i['Time Delays 2'], dfO1_f['Time Delays 3'], 
                  dfO2_f['Time Delays 4']], axis=1)
dfNS3 = dfNS3.rename(columns={"Time Delays 1": "Initial; O1", "Time Delays 2": "Initial; O2", "Time Delays 3": "Post; O1", 
                            "Time Delays 4": "Post; O2"}, errors="raise")
dfNS3
```
