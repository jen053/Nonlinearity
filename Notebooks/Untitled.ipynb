```python
import numpy as np
import seaborn as sns
import math as mt
import matplotlib
import matplotlib.pyplot as pl
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
pl.rcParams["font.family"] = "Arial"
pl.rcParams["font.size"] = 16
import matplotlib.colors as mcolors
```

```python
# Parameter (reference height)
h = 0.0

# Code to generate the data
n = 300
alpha = 0.75 * mt.pi
theta = [alpha + 2.0 * mt.pi * (float(k) / float(n)) for k in range(0, n + 1)]
xs = [k+i for k, i in zip(theta, np.arange(0, 1000, 1))]
ys = [(1/2) * mt.sin(2.5*k) for k in theta]
zs = [k for k in np.ones_like(theta)]


v = []
for k in range(0, len(xs) - 1):
    x = [xs[k], xs[k+1], xs[k+1], xs[k]]
    y = [ys[k], ys[k+1], ys[k+1], ys[k]]
    z = [zs[k], zs[k+1],       h,     h]
    v.append(list(zip(x, y, z))) 
poly3dCollection = Poly3DCollection(v, color = mcolors.TABLEAU_COLORS['tab:blue'], alpha=0.3)


xs1 = [k+i for k, i in zip(theta, np.arange(0, 1000, 1))]
ys1 = [(1/2) * mt.sin(2.5*k) +4 for k in theta]
zs1 = [k for k in np.ones_like(theta)]
v1 = []
for k in range(0, len(xs1) - 1):
    x1 = [xs1[k], xs1[k+1], xs1[k+1], xs1[k]]
    y1 = [ys1[k], ys1[k+1], ys1[k+1], ys1[k]]
    z1 = [zs1[k], zs1[k+1],       h1,     h1]
    v1.append(list(zip(x1, y1, z1))) 
poly3dCollection1 = Poly3DCollection(v1, color = mcolors.TABLEAU_COLORS['tab:blue'], alpha=0.3)


xs2 = [k+i for k, i in zip(theta, np.arange(0, 1000, 1))]
ys2 = [(1/2) * mt.sin(2.5*k) -4 for k in theta]
zs2 = [k for k in np.ones_like(theta)]
v2 = []
for k in range(0, len(xs1) - 1):
    x2 = [xs2[k], xs2[k+1], xs2[k+1], xs2[k]]
    y2 = [ys2[k], ys2[k+1], ys2[k+1], ys2[k]]
    z2 = [zs2[k], zs2[k+1],       h,     h]
    v2.append(list(zip(x2, y2, z2))) 
poly3dCollection2 = Poly3DCollection(v2, color = mcolors.TABLEAU_COLORS['tab:blue'], alpha=0.3)


# Code to plot the 3D polygons
fig = pl.figure(figsize=(10,10))
ax = Axes3D(fig)
ax.add_collection3d(poly3dCollection)
ax.add_collection3d(poly3dCollection1)
ax.add_collection3d(poly3dCollection2)
ax.set_xlim([min(xs), max(xs)])
ax.set_ylim([-5, 5])
ax.set_zlim([0, max(zs)])
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_axis_off()
ax.view_init(30,210)
pl.show()
```
