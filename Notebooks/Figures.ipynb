```python
import matplotlib
import numpy as np

font = {'family' : 'Times New Roman',
        'weight' : 'bold',
        'size'   : 16}

matplotlib.rc('font', **font)
import matplotlib.style as style
style.use('seaborn-paper')
import matplotlib.pyplot as plt
```

```python
stress = np.arange(0, 200, 0.1)
```

```python
def linearElastic(x, m, b):
    return m*x + b
```

```python
strainLin = linearElastic(stress, 1.5, 0)
```

```python
def nonlinearElastic(x, a, b):
    return a*x**b
```

```python
strainNonlin = nonlinearElastic(stress, 1, 1.35)
```

```python
def hysteresis_x(m, n, a, bx, alpha):
    return a * (np.cos(alpha))**m + bx * (np.sin(alpha))**n

def hysteresis_y(by, alpha):
    return by* np.sin(alpha)
```

```python
hysx = hysteresis_x(3, 1, 0.2, 1, np.arange(0,10,0.01))
hysy = hysteresis_y(1, np.arange(0,10,0.01) )
```

```python
fig, ax = plt.subplots(1, 3, figsize = (10,5))
ax[0].plot(strainLin, stress,  linewidth=2.5, color = 'navy')
ax[1].plot(strainNonlin, stress, linewidth=2.5, color = 'green')
ax[2].plot(hysx, hysy,  linewidth=2.5, color = 'orange')

[ax[i].grid() for i in np.arange(0, 3 ,1)]
[ax[i].set_xlabel('Strain', fontsize=16) for i in np.arange(0, 2 ,1)]
[ax[i].set_ylabel('Stress', fontsize=16) for i in np.arange(0, 2 ,1)]
ax[2].set_xlabel('x(stress)', fontsize=16)
ax[2].set_ylabel('y(stress)', fontsize=16)
plt.tight_layout()
[ax[i].axes.xaxis.set_ticks([]) for i in np.arange(0,3,1)]
[ax[i].axes.yaxis.set_ticks([]) for i in np.arange(0,3,1)]
plt.show()
```
